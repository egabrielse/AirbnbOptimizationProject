```julia
using JuMP, CSV, DataFrames, Gurobi, PyPlot
listings = CSV.read("listings_bos.csv");
```

```julia
price = listings[:,9]
latitude = listings[:,3]
longitude = listings[:,4];
```

```julia
model = Model(Gurobi.Optimizer)
set_optimizer_attribute(model,"OutputFlag",false)

@variable(model, x[1:3585] >= 0)

@objective(model, Min, sum(x[i] * price[i] for i in 1:3585))

@constraint(model, sum(x[i] for i in 1:3585) == 4)
@constraint(model, book_once[i in 1:3585], x[i] <= 1)

optimize!(model)

longitude_optimal = []
latitude_optimal = []

println("Min Price: ", objective_value(model))
for i in 1:3585
    if value(x[i]) > 0
        print(i, ":", value(x[i]), " ")
        append!(longitude_optimal, longitude[i])
        append!(latitude_optimal, latitude[i])
    end
end
```

```julia
# plot the points
using PyPlot
figure()
plot(longitude, latitude, "b.", markersize=1)
plot(longitude_optimal, latitude_optimal, "r.", markersize=10)
title("Airbnb Locations")
xlabel("Longitude")
ylabel("Latitude");
grid()
```
